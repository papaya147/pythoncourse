```python
from sklearn.datasets import fetch_california_housing
housing_data = fetch_california_housing()
```

```python
housing = DataFrame(housing_data.data, columns = housing_data.feature_names)
housing.head()
```

```python
corr = housing.corr()
corr
```

```python
corr["MedInc"]["AveRooms"] # gives corelation between specific columns
```

```python
import seaborn as sns # better matrix visuals

cmap = sns.diverging_palette(10,255,as_cmap = True)
sns.heatmap(corr, cmap = cmap)
```

```python
from pandas.plotting import scatter_matrix # main diagonal has been repurposed to show 
                                           # distribution of data (corr of it is always 1)

a = scatter_matrix(housing,figsize = (16,16))
```
